## **Merging DataFrame**

```python
import pandas as pd
```

```python
foods = pd.read_csv("restaurant_foods.csv")
```

```python
customers = pd.read_csv("restaurant_customers.csv")
```

```python
week1 = pd.read_csv("restaurant_week_1_sales.csv")
```

```python
week2 = pd.read_csv("restaurant_week_2_sales.csv")
```

```python
len(week1)
```

```python
len(week2)
```

```python
pd.concat([week1,week2])
```

```python
pd.concat([week1,week2], keys = ["Week_1","Week_2"]).sort_index()
```

```python
week1.merge(foods, how = "left", on = "Food ID")
```

```python
week2.merge(customers, left_on= "Customer ID", right_on = "ID", how = "left").drop("ID", axis = "columns")
```

```python
week1.merge(week2, how = "inner", on = "Customer ID", suffixes=[" - Week 1", " - Week2"])
```

```python
week1.merge(week2, how = "inner", on = ["Customer ID", "Food ID"])
```

```python
merged = week1.merge(week2, how = "outer", on = "Customer ID", suffixes=[" - Week_1 ", " - Week_2"], indicator=True)
```

```python
merged["_merge"].value_counts()
```

```python
merged[merged["_merge"].isin(["left_only", "right_only"])]
```

```python
foods = pd.read_csv("restaurant_foods.csv", index_col="Food ID")
customer = pd.read_csv("restaurant_customers.csv", index_col="ID")
```

```python
week1.merge(customer, how = "left", left_on = "Customer ID", right_index=True
            ).merge(foods, how = "left", left_on = "Food ID", right_index= True)
```

```python
times = pd.read_csv("restaurant_week_1_times.csv")
```

```python
times.head()
```

```python
week1.merge(times, how = "left", left_index = True, right_index = True)
```

```python
week1.join(times)
```
